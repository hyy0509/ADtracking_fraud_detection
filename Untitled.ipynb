```python
def lgb_modelfit_nocv(params, dtrain, dvalid, predictors, target='target', objective='binary', metrics='auc',
                 feval=None, early_stopping_rounds=20, num_boost_round=3000, verbose_eval=10, categorical_features=None):
    lgb_params = {
        'boosting_type': 'gbdt',
        'objective': objective,
        'metric':metrics,
        'learning_rate': 0.2,
        #'is_unbalance': 'true',  #because training data is unbalance (replaced with scale_pos_weight)
        'num_leaves': 31,  # we should let it be smaller than 2^(max_depth)
        'max_depth': -1,  # -1 means no limit
        'min_child_samples': 20,  # Minimum number of data need in a child(min_data_in_leaf)
        'max_bin': 255,  # Number of bucketed bin for feature values
        'subsample': 0.6,  # Subsample ratio of the training instance.
        'subsample_freq': 0,  # frequence of subsample, <=0 means no enable
        'colsample_bytree': 0.3,  # Subsample ratio of columns when constructing each tree.
        'min_child_weight': 5,  # Minimum sum of instance weight(hessian) needed in a child(leaf)
        'subsample_for_bin': 200000,  # Number of samples for constructing bin
        'min_split_gain': 0,  # lambda_l1, lambda_l2 and min_gain_to_split to regularization
        'reg_alpha': 0,  # L1 regularization term on weights
        'reg_lambda': 0,  # L2 regularization term on weights
        'nthread': 4,
        'verbose': 0,
        'metric':metrics
    }

    lgb_params.update(params)

    print("preparing validation datasets")

    xgtrain = lgb.Dataset(dtrain[predictors].values, label=dtrain[target].values,
                          feature_name=predictors,
                          categorical_feature=categorical_features
                          )
    xgvalid = lgb.Dataset(dvalid[predictors].values, label=dvalid[target].values,
                          feature_name=predictors,
                          categorical_feature=categorical_features
                          )

    evals_results = {}

    bst1 = lgb.train(lgb_params, 
                     xgtrain, 
                     valid_sets=[xgtrain, xgvalid], 
                     valid_names=['train','valid'], 
                     evals_result=evals_results, 
                     num_boost_round=num_boost_round,
                     early_stopping_rounds=early_stopping_rounds,
                     verbose_eval=10, 
                     feval=feval)

    print("\nModel Report")
    print("bst1.best_iteration: ", bst1.best_iteration)
    print(metrics+":", evals_results['valid'][metrics][bst1.best_iteration-1])

    return (bst1,bst1.best_iteration)
```

```python
# feature engineering
def fcount(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-1])[selcols[-1]].\
    count().reset_index().rename(index=str, columns={selcols[len(selcols)-1]:fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-1],how='left')
    return df
def fmean(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-1])[selcols[-1]].\
    mean().reset_index().rename(index=str, columns={selcols[len(selcols)-1]:fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-1],how='left')
    return df
def fvar(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-1])[selcols[-1]].\
    var().reset_index().rename(index=str, columns={selcols[len(selcols)-1]:fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-1],how='left')
    return df
def fskew(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-1])[selcols[-1]].\
    skew().reset_index().rename(index=str, columns={selcols[len(selcols)-1]:fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-1],how='left')
    return df
def fnunique(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-1])[selcols[-1]].\
    nunique().reset_index().rename(index=str, columns={selcols[len(selcols)-1]:fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-1],how='left')
    return df
def fcumcount(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-1])[selcols[-1]].\
    cumcount()
    df[fname] = gp.values
    return df
def fcount_group(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-2])[selcols[-1]].\
    count().reset_index().rename(index=str, columns={selcols[-1]: fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-2], how='left')
    return df
def fcount_group(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-2])[selcols[-1]].\
    count().reset_index().rename(index=str, columns={selcols[-1]: fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-2], how='left')
    return df
def fmean_group(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-2])[selcols[-1]].\
    mean().reset_index().rename(index=str, columns={selcols[-1]: fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-2], how='left')
    return df
def fvar_group(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-2])[selcols[-1]].\
    var().reset_index().rename(index=str, columns={selcols[-1]: fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-2], how='left')
    return df
def fcumcount_group(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-2])[selcols[-1]].\
    cumcount().reset_index().rename(index=str, columns={selcols[-1]: fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-2], how='left')
    return df
def fmean_group(df,selcols,fname):
    gp = df[selcols].groupby(by=selcols[0:len(selcols)-2])[selcols[-1]].\
    mean().reset_index().rename(index=str, columns={selcols[-1]: fname})
    df = df.merge(gp, on=selcols[0:len(selcols)-2], how='left')
    return df
```

```python
# add features ip	app	device	os	channel	click_time	attributed_time	is_attributed	hour	day
columns = ['ip', 'app', 'device', 'os', 'channel', 'click_time', 'is_attributed', 'hour', 'day']

for i in range(len(columns)-1):
    selcols = ['day']
    if columns[i] == 'is_attributed':
        selcols.append(columns[i])
        train = fcount(train, selcols, 'c_'+columns[i])
        train = fmean(train, selcols, 'm_'+columns[i])
        train = fvar(train, selcols, 'v_'+columns[i])
        train = fskew(train, selcols, 'sk_'+columns[i])
        train = fnunique(train, selcols, 'nuni_'+columns[i])
        train = fcumcount(train, selcols, 'cum_c_'+columns[i])
    else:
        selcols.append(columns[i])
        train = fcount(train, selcols, 'c_'+columns[i])
        train = fmean(train, selcols, 'm_'+columns[i])
        train = fvar(train, selcols, 'v_'+columns[i])
        train = fskew(train, selcols, 'sk_'+columns[i])
        train = fnunique(train, selcols, 'nuni_'+columns[i])
        train = fcumcount(train, selcols, 'cum_c_'+columns[i])
        test = fcount(test, selcols, 'c_'+columns[i])
        test = fmean(test, selcols, 'm_'+columns[i])
        test = fvar(test, selcols, 'v_'+columns[i])
        test = fskew(test, selcols, 'sk_'+columns[i])
        test = fnunique(test, selcols, 'nuni_'+columns[i])
        test = fcumcount(test, selcols, 'cum_c_'+columns[i])
```
